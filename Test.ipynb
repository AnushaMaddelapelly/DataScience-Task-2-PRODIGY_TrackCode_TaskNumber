```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#To print first 5 rows of data
df=pd.read_csv('Task2test.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
#discription of dataset
df.describe(include="all")
```

```python
df.info()
```

```python
#to count number of females in test data
female_count=len(df[df['Sex'] == 'female'])
print("Number of females in test data:",female_count)
```

```python
#to coount number of males in test data
male_count=len(df[df['Sex'] == 'male'])
print("Number of males in test data:",male_count)
```

```python
#female and male count plotting
categories=['female','male']
counts=[152,266]
plt.bar(categories,counts,color=['red','blue'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('female vs male count')
plt.show()
```

```python
import seaborn as sns
sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize=(7,5))
sns.histplot(df['Fare'], bins=25, kde=True, color='blue', edgecolor='black', ax=ax)
ax.set_title('Distribution of Fare', fontsize=16,fontweight='bold')
ax.set_xlabel('Fare')
ax.set_ylabel('Frequency', fontsize=12)
ax.grid(True, linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
class_counts = df['Pclass'].value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.bar(class_counts.index,class_counts.values,color='lightgreen',edgecolor='black')
plt.title('passengers count by class')
plt.xlabel('count')
plt.ylabel('Passenger Class')
plt.grid(True, axis='x')
plt.show()
```

```python
plt.hist(df['SibSp'], bins=10, color='red', edgecolor='black')
plt.title('Distribution of SibSp')
plt.show()
```

```python
df['Parch'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Distribution of Parch')
plt.xlabel('Parch')
plt.ylabel('Frequency')
plt.show()
```
